```r
#if and else statements
# Define a variable 
x <- 5
# Use if-else statement
if (x > 0) {
  print("x is positive")
} else if (x < 0) {
  print("x is negative")
} else {
  print("x is zero")
}
```

```r
x <- 5
# Use if-else statement
if (x > 4) {
  print("x is greater than 4")
} else if (x < 0) {
  print("x is smaller than 4")
} else {
  print("x is equal to 4")
}
```

```r
#nested if-else statements
x <- 5
# Use nested if-else statement
if (x > 0) {
    print("x is positive")
    if (x > 10) {
        print("x is greater than 10")
    } else {
        print("x is less than or equal to 10")
    }
    } else if (x < 0) {
    print("x is negative")
    } else {
    print("x is zero")
    }

```

```r
#creating marks division
marks <- 85
if (marks >= 90) {
  print("Grade: A")
} else if (marks >= 80) {
  print("Grade: B")
} else if (marks >= 70) {
  print("Grade: C")
} else if (marks >= 60) {
  print("Grade: D")
} else {
  print("Grade: F")
}
```

```r
#PRINT NUMBER IN REVERSE ORDER USING WHILE
num <- 10
while (num >= 1) {
  print(num)
  num <- num - 1
}
```

```r
#WRITE A WHILE LOOPTO CALULATE THE SUM OF FIRST 10 NATURAL NUMBERS
sum <- 0
i <- 1
while (i <= 10) {
  sum <- sum + i
  i <- i + 1
}
print(paste("The sum of the first 10 natural numbers is:", sum))
```

```r
#WRITE A WHILE LOOP TO PRINT EVEN NUMBERS FROM 1 TO 20
even_sum <- 0
i <- 1
while (i <= 20) {
    if (i %% 2 == 0) {
        even_sum <- even_sum + i
    }
    i <- i + 1
    }
print(paste("The sum of even numbers from 1 to 20 is:", even_sum))
```

```r
#WRITE A WHILE LOOP TO PRINT EVEN NUMBERS FROM 1 TO 20 WITHOUT IF
num <- 2
while (num <= 20) {
  print(num)
  num <- num + 2
}
```
